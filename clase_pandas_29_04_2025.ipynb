```python
from math import *
import numpy as np
import pandas as pd


def g(x):
    return cos(x)


def punto_fijo(g, n , x):
    i = 1
    while i < n:
        x = g(x)
        i = i+1
    return x


sucesion = np.array([punto_fijo(g, n, 1) for n in range(1, 50)])

#indice = range(1, 10)

#mi_dataframe = pd.DataFrame(sucesion, columns=[ 'xn'])

mi_dataframe = pd.DataFrame(sucesion, columns=[ 'xn'], index=range(1, 50))



mi_dataframe

```

```python
mi_dataframe['shift'] = mi_dataframe['xn'].shift(-1)

mi_dataframe
```

```python
mi_dataframe['cociente'] = mi_dataframe['shift']/mi_dataframe['xn']

mi_dataframe
```
